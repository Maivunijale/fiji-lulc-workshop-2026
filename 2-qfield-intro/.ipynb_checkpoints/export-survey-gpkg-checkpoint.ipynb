```python
from shapely.geometry import Point
import geopandas as gpd
```

#### Insert markdown instructions / explanation

```python
points = [
        Point(178.4591312, -18.1212562), 
        Point(178.4612535, -18.1217040),  
        Point(178.4568836, -18.1194217),
]

len_data = len(points)
len_data
```

#### Insert markdown instructions / explanation

```python
data = {
    'station_id': list(range(1, len_data+1)),
    'type': [objective] * len_data, 
    'observer': [surveyor] * len_data,
    'geometry': [
        Point(178.4591312, -18.1212562), 
        Point(178.4612535, -18.1217040),  
        Point(178.4568836, -18.1194217),
    ]
}
```

```python
gdf = gpd.GeoDataFrame(data, crs="EPSG:4326")
```

#### Insert markdown instructions / explanation

```python
gdf.explore(
    marker_type='circle_marker', 
    marker_kwds={
        'radius': 10,           
        'fill': True,
        'color': 'red'          
    }
)
```

```python
gdf.to_file("field_data.gpkg", layer="survey_layer", driver="GPKG")
```
